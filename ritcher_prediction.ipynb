## Part 1 - Data Preprocessing

```python
train=pd.read_csv("new_train.csv")
train.drop(columns=["Unnamed: 0"],inplace=True)
train
```

```python
arr=train.columns.values
print(np.where(arr=="legal_ownership_status"))
```

```python
X=train.iloc[:,:-1]
y=train.iloc[:,-1]
```

```python
X
```

```python
X.columns.values
```

```python
X_cat=X[["land_surface_condition","foundation_type",'roof_type','ground_floor_type','other_floor_type','plan_configuration','position','legal_ownership_status']].copy()
X_cat
```

```python
X_ids=X[['building_id', 'geo_level_1_id', 'geo_level_2_id',
       'geo_level_3_id']].copy()
X_ids
```

```python
X_rem=X.drop(columns=['building_id', 'geo_level_1_id', 'geo_level_2_id',
       'geo_level_3_id',"land_surface_condition","foundation_type",'roof_type','position','ground_floor_type','other_floor_type','plan_configuration','legal_ownership_status'])
X_rem
```

**We need to use one-hot encoding for categorical data**
land_surface_condition,foundation_type,roof_type,ground_floor_type,other_floor_type,position,plan_configuration
,position,plan_configuration

```python
from sklearn.preprocessing import OneHotEncoder
X_cat=OneHotEncoder().fit_transform(X_cat).toarray()
```

### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_rem=sc.fit_transform(X_rem)
```

### Merging all the three different into one

```python
X=np.concatenate((X_ids,X_rem,X_cat),axis=1)
X.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

## Training and testing data

**Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression(max_iter=20000)
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

**RandomForestClassifier**

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
y_pred=model.predict(X_test)
```

**Decision Tree**

```python
from sklearn import tree
model=tree.DecisionTreeClassifier()
model.fit(X_train,y_train)
```

```python
 model.score(X_test,y_test)
```

### The predicted data of Random Forest Classifier is giving the best accuracy
